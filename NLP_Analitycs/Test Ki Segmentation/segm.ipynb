```python
df_contenido=df[["Contenido"]]
# df_contenido=df_contenido.head()
df_contenido
```

```python
def segmentacion(cont):
    lista=[]
    nlp=spacy.load("es_core_news_sm")
    doc=nlp(cont)
    for sent in doc.sents:
        lista.append(sent.text)
    return lista
```

```python
df1 = df_contenido["Contenido"].apply(segmentacion)
```
